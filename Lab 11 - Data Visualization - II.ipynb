```python
import seaborn as sns
```

```python
tips_data = sns.load_dataset('tips')
```

```python
tips_data.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips_data)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips_data, color='red')
```

```python
sns.set_palette('husl')

sns.scatterplot(x='total_bill', y='tip', data=tips_data)
```

```python
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data=tips_data)
plt.title('Total bill vs tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
```

```python
sns.lineplot(x='day', y='tip', data=tips_data)
```

```python
sns.regplot(x='total_bill', y='tip', data=tips_data)
```

```python
sns.barplot(x='day', y='total_bill', data=tips_data)
```

```python
sns.histplot(x='total_bill', data=tips_data)
```

```python
correlation = tips_data.corr()
correlation
```

```python
sns.heatmap(correlation, annot=True, cmap='coolwarm')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips_data)
```

```python
sns.countplot(x='day', data=tips_data)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips_data)
```

```python
sns.pairplot(tips_data)
```
