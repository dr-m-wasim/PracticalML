```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = 2 * np.random.randn(100,1) + 3
```

```python
x[:10]
```

```python
y = 256 + 6 * x + np.random.randn(100,1)
```

```python
y[:10]
```

```python
plt.plot(x, y, 'b^')
plt.xlabel('x')
plt.ylabel('y')
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(x, y)
```

```python
print(model.intercept_, model.coef_)
```

```python
new_house = 5
m = model.coef_
c = model.intercept_
```

```python
m * new_house + c
```

```python
model.predict([[5]])
```

```python
line = np.linspace(0, 10, 100).reshape(-1, 1)
```

```python
predictions = model.predict(line)
```

```python
plt.scatter(x, y)
plt.plot(line, predictions, c='red')
```
